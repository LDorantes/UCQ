```python
import pandas as pd

# Cargar los archivos CSV
file_path_adaptive = "Adaptive_AI_Feedback_Dataset.csv"
file_path_cognitive = "cognitive.csv"
file_path_User_Interactions_Dataset = "userInteraction.csv"


df_adaptive = pd.read_csv(file_path_adaptive)
df_cognitive = pd.read_csv(file_path_cognitive)
df_UI = pd.read_csv(file_path_User_Interactions_Dataset)

# Mostrar las primeras filas de cada dataset para entender su estructura
df_adaptive.head(), df_cognitive.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo
sns.set(style="whitegrid")

# Gráfico de ajustes de dificultad
plt.figure(figsize=(8, 5))
sns.histplot(df_adaptive["difficulty_adjustment"], bins=5, kde=True, color="blue")
plt.title("Distribución de Ajustes de Dificultad")
plt.xlabel("Ajuste de Dificultad")
plt.ylabel("Frecuencia")
plt.show()

# Gráfico de niveles de concentración (focus_level)
plt.figure(figsize=(8, 5))
sns.histplot(df_cognitive["focus_level"], bins=20, kde=True, color="green")
plt.title("Distribución del Nivel de Concentración")
plt.xlabel("Nivel de Concentración")
plt.ylabel("Frecuencia")
plt.show()

# Relación entre rendimiento y ajustes de dificultad
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_adaptive, x="player_performance", y="difficulty_adjustment", hue="challenge_rating", palette="coolwarm")
plt.title("Relación entre Rendimiento del Jugador y Ajuste de Dificultad")
plt.xlabel("Rendimiento del Jugador")
plt.ylabel("Ajuste de Dificultad")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datasets
df_interactions = pd.read_csv("userInteraction.csv")
df_cognitive = pd.read_csv("cognitive.csv")
df_feedback = pd.read_csv("Adaptive_AI_Feedback_Dataset.csv")

# Configuración de estilo
sns.set(style="whitegrid")

# --- 1. User Interactions Dataset: Distribución del Score ---
plt.figure(figsize=(8, 5))
sns.histplot(df_interactions['score'], bins=20, kde=True, color="blue")
plt.title("Distribución de Puntos Obtenidos en las Sesiones")
plt.xlabel("Puntaje")
plt.ylabel("Frecuencia")
plt.show()

# --- 2. Cognitive States Dataset: Niveles de Concentración y Estrés ---
plt.figure(figsize=(8, 5))
sns.boxplot(data=df_cognitive[['focus_level', 'stress_level']], palette="Set2")
plt.title("Distribución de Niveles de Concentración y Estrés")
plt.xlabel("Métrica")
plt.ylabel("Nivel (0-100)")
plt.show()

# --- 3. Adaptive AI Feedback Dataset: Ajustes de Dificultad ---
plt.figure(figsize=(8, 5))
sns.countplot(x=df_feedback['difficulty_adjustment'], palette="coolwarm")
plt.title("Frecuencia de Ajustes de Dificultad por la IA")
plt.xlabel("Ajuste de Dificultad")
plt.ylabel("Cantidad de Ajustes")
plt.show()

# --- Extra: Relación entre Fatiga y Estado de Flujo ---
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df_cognitive['fatigue_level'], y=df_cognitive['flow_state_score'], alpha=0.6, color="purple")
plt.title("Relación entre Fatiga y Estado de Flujo")
plt.xlabel("Nivel de Fatiga")
plt.ylabel("Estado de Flujo")
plt.show()
```
